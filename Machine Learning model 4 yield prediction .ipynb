```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import seaborn as sns
sns.set()
```

```python
#Feature selection#
def plot_correlation_map( df ):
    
    corr = df.corr()

    s , ax = plt.subplots( figsize =( 14 , 14) )

    cmap = sb.diverging_palette( 220 , 10 , as_cmap = True )

    s = sb.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 9 }

        )
```

```python
data.corr()
```

```python
plot_correlation_map(data)
```

**SVR (Support Vector Regression)**

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
x3=yield_data[["respondent_age","n_crops_produced","fm_cocoa_lab_n_ha","cocoa_manurecomp_yn","cocoa_pest_yn","cocoa_training_5years_yn"]]
#target_extraction
y3=yield_data[["cocoa_prod_total_kgsha"]]
```

```python
# prédictions sur l'ensemble de test
predict3 = model3.predict(X_test_scaled)

```

```python
# Évaluation de la performance du modèle (avec l'erreur quadratique moyenne)
mse = mean_squared_error(y_test3, predict3)
print(f'Mean Squared Error: {mse}')
```

**AdaBoostRegressor**

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x5=yield_data[["respondent_age","n_crops_produced","fm_cocoa_lab_n_ha","cocoa_manurecomp_yn","cocoa_pest_yn","cocoa_training_5years_yn"]]
#target_extraction
y5=yield_data[["cocoa_prod_total_kgsha"]]


# ensembles d'entraînement et de test
X_train5, X_test5, y_train5, y_test5 = train_test_split(x5, y5, test_size=0.3, random_state=42)

#  modèle AdaBoostRegressor avec un arbre de décision comme estimateur de base
base_model5 = DecisionTreeRegressor(max_depth=3)  # Vous pouvez ajuster max_depth selon vos besoins
model5 = AdaBoostRegressor(base_model5, n_estimators=50, random_state=42)

# Entraînez le modèle sur l'ensemble d'entraînement
model5.fit(X_train5, y_train5)

# Faites des prédictions sur l'ensemble de test
predict5 = model5.predict(X_test5)

# Évaluez la performance du modèle (par exemple, avec l'erreur quadratique moyenne)
mse = mean_squared_error(y_test5, predict5)
print(f'Mean Squared Error: {mse}')
```

**BaggingRegressor**

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


x9=yield_data[["respondent_age","n_crops_produced","fm_cocoa_lab_n_ha","cocoa_manurecomp_yn","cocoa_pest_yn","cocoa_training_5years_yn"]]
#target_extraction
y9=yield_data[["cocoa_prod_total_kgsha"]]

X_train9, X_test9, y_train9, y_test9 = train_test_split(x9, y9, test_size=0.3, random_state=42)

base_model9 = DecisionTreeRegressor(max_depth=3)  # Vous pouvez ajuster max_depth selon vos besoins
model9 = BaggingRegressor(base_model9, n_estimators=50, random_state=42)

model9.fit(X_train9, y_train9)

# Faites des prédictions sur l'ensemble de test
predict9 = model9.predict(X_test9)

# Évaluez la performance du modèle (par exemple, avec l'erreur quadratique moyenne)
mse = mean_squared_error(y_test9, predict9)
print(f'Mean Squared Error: {mse}')
```
