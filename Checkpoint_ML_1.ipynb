# Feature selection

```python
def plot_correlation_map( df ):
    
    corr = df.corr()

    s , ax = plt.subplots( figsize =( 14 , 14) )

    cmap = sb.diverging_palette( 220 , 10 , as_cmap = True )

    s = sb.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 9 }

        )
```

```python
house.corr()
```

```python
plot_correlation_map(house)
```
